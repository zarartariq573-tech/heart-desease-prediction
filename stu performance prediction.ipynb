```python
x = data.drop("math score", axis=1)
y = data["math score"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
from catboost import CatBoostRegressor
from sklearn.model_selection import RandomizedSearchCV, train_test_split
```

```python
cat_features = [0,1,2,3,4]
```

```python
model = CatBoostRegressor(  iterations=700,           # enough boosting rounds
    learning_rate=0.05,        # small step size
    depth=5,                   # balanced tree depth
    l2_leaf_reg=8,             # prevents overfitting
    bagging_temperature=1,     # adds randomness
    subsample=0.8,             # only 80% samples per tree
    colsample_bylevel=0.8,     # only 80% features per split
    random_seed=42,
    verbose=100,
    early_stopping_rounds=100)
model.fit(x_train,y_train,cat_features=cat_features)
```

```python
model.score(x_test,y_test)*100
```

```python
model.predict([['female',	'group C',	'high school',	'free/reduced',	'completed', 71,	65]])
```
